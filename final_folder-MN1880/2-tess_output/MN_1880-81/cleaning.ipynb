```python
import re
lines_beg_cleaned = []
for i in range(len(lines)-1):
    if(len(lines[i+1]) < 15):
        lines_beg_cleaned.append(re.sub(r"^\W+", "", lines[i])[:-2] + ' ' + lines[i+1])
    elif(len(lines[i]) >= 15):
        lines_beg_cleaned.append(re.sub(r"^\W+", "", lines[i]))
if(i == len(lines)-2):
        lines_beg_cleaned.append(re.sub(r"^\W+", "", lines[len(lines)-1]))

lines_beg_cleaned
```

```python
with open(r"MN_1880-81_cleaned.txt","w",encoding = "utf-8") as fout:
    fout.writelines(lines_beg_cleaned)
```

```python
def contains_number(s):
    return any(i.isdigit() for i in s)
```

```python
splitted =[]
for line in lines_beg_cleaned:
    splitted.append(re.split('(?<!\d)[,.]|[,.](?!\d)', line))
for line in splitted:
    for entry in line:
        if(contains_number(entry) and len(entry.split()) >= 4 and len(line)>7):
            new_entry_start = entry.split()[-2]
            print(line, ':',new_entry_start)
        
```
