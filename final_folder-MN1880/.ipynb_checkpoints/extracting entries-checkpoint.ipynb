```python
import json 
import re 
import os
from pathlib import Path
```

```python
pathlist = Path('tess_output\\MN_1850-51').glob('**/*.json')
for path in pathlist:
        with open(str(path), encoding="utf8") as json_file: 
            print(str(json_file))
            data = json.load(json_file)
            target_entity = data['pages'][0]['lines']
            for i in range(len(target_entity)):
                if('completeText' in list(target_entity[i].keys())):
                    sentence = target_entity[i]['completeText']
                    str2 = sentence.replace("\n", "")
                    address = re.sub(' +', ' ',str2)
                    with open("tess_output/MN_1850-51/textoutput.txt", "a", encoding="utf8") as f:
                        print(address)
                        f.write(address)
                        f.write("\n")
```
