```python
import json 
import re 
import os
from pathlib import Path
```

```python
pathlist = Path('').glob('**/*.json')
for path in pathlist:
        with open(str(path), encoding="utf8") as json_file: 
#             print(str(json_file))
            data = json.load(json_file)
            target_entity = data['pages'][0]['lines']
            for i in range(len(target_entity)):
                if('text' in list(target_entity[i].keys())):
                    sentence = target_entity[i]['text']
                    str2 = sentence.replace("\n", "")
                    address = re.sub(' +', ' ',str2)
                    with open("textoutput.txt", "a", encoding="utf8") as f:
                        f.write(address)
                        f.write("\n")
```

```python
#cleaning
with open(r"textoutput.txt","r",encoding = "utf-8") as f:
    lines = f.readlines()
```

```python
import re
lines_beg_cleaned = []
for line in lines:
    lines_beg_cleaned.append(re.sub(r"^\W+", "", line))
lines_beg_cleaned
```

```python
final_lines = list()
final_lines.append(lines_beg_cleaned[0])
for i in range(1,len(lines_beg_cleaned)):
    if lines_beg_cleaned[i] not in lines_beg_cleaned[i-1]:
        final_lines.append(lines_beg_cleaned[i])
final_lines
```

```python
final_lines = [item for item in final_lines if len(item)>=4]
```

```python
with open(r"BK_1880_cleaned.txt","w",encoding = "utf-8") as fout:
    fout.writelines(final_lines)
```
